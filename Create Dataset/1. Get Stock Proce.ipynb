```python
import pandas as pd
import yfinance as yf
```

## Get the names of the S&P 500 constituent stocks from Wikipedia

```python
sp500url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
data_table = pd.read_html(sp500url)
```

## Then import the stock prices through Yahoo Finance based on the list of stock names

```python
snp_prices = yf.download(tickers)
```

#### After check Weki, we find that BRK.B is Berkshire Hathaway Inc. and BF.B is Brown-Forman Corporation, they use - instead of . in their tickers

```python
yf.download('BF-B')['Adj Close']
```

```python
yf.download('BRk-B')['Adj Close']
```

```python
for i in range(len(tickers)):
    if tickers[i] == 'BF.B':
        tickers[i] = 'BF-B'
    elif tickers[i] == 'BRK.B':
        tickers[i] = 'BRK-B'
```

```python
'BF-B' in tickers
'BRK-B' in tickers
```

## Change the code to whatever period you want

```python
snp_prices = yf.download(tickers, start='2014-06-15', end='2024-06-15')['Adj Close']
```

```python
snp_prices.to_csv('snp_prices.csv')
```
